```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pickle
import glob
import pandas as pd
```

```python
rundir = "../models/run01_hz/"
```

```python
hurdle_prob_train_mae = []
hurdle_prob_test_mae = []
hurdle_prob_train_rmse = []
hurdle_prob_test_rmse = []
```

```python
nfolds = 5
for fold in np.arange(1,nfolds+1):
    hurdle_res = pickle.load(open(rundir+str(fold)+"/results_hurdle.pickle","rb"))
    
    hurdle_prob_train_mae.append(hurdle_res['train_hurdle_comb_mae'])
    hurdle_prob_test_mae.append(hurdle_res['test_hurdle_comb_mae'])
    hurdle_prob_train_rmse.append(hurdle_res['train_hurdle_comb_rmse'])
    hurdle_prob_test_rmse.append(hurdle_res['test_hurdle_comb_rmse'])

hurdle_prob_train_mae = np.hstack(hurdle_prob_train_mae)
hurdle_prob_test_mae = np.hstack(hurdle_prob_test_mae)
hurdle_prob_train_rmse = np.hstack(hurdle_prob_train_rmse)
hurdle_prob_test_rmse = np.hstack(hurdle_prob_test_rmse)
```

```python
print("hurdle prob - mean train mae : "+str(np.round(hurdle_prob_train_mae.mean(),4))+" (1sigma: "+str(np.round(hurdle_prob_train_mae.std(),4))+")")
print("hurdle prob - mean test  mae : "+str(np.round(hurdle_prob_test_mae.mean(),4))+" (1sigma: "+str(np.round(hurdle_prob_test_mae.std(),4))+")")
```

```python
print("hurdle prob - mean train rmse : "+str(np.round(hurdle_prob_train_rmse.mean(),4))+" (1sigma: "+str(np.round(hurdle_prob_train_rmse.std(),4))+")")
print("hurdle prob - mean test  rmse : "+str(np.round(hurdle_prob_test_rmse.mean(),4))+" (1sigma: "+str(np.round(hurdle_prob_test_rmse.std(),4))+")")
```

** visualization **

```python
rundir = "../models/run01/1/"
data = pickle.load(open(rundir+"data.pickle","rb"))
hurdle_results = pickle.load(open(rundir+"results_hurdle.pickle","rb"))

Xtrain = data['Xtrain']
Ytrain = data['Ytrain']
Xtest = data['Xtest']
Ytest = data['Ytest']

traindf = pd.DataFrame({'ndatehour':Xtrain[:,2].flatten()*1000,'pptr':Ytrain.flatten(),'hurdle':hurdle_results['train_pred_hurdle_comb'].flatten()})
mpl.rcParams['figure.figsize'] = (16,4)
fig, ax = plt.subplots(1, 1, sharex=True)

mean_pptr = traindf.groupby('ndatehour')['pptr'].mean()
ax.bar(mean_pptr.index, mean_pptr.values, align='center')
ax2 = ax.twinx()
mean_hurdle = traindf.groupby('ndatehour')['hurdle'].mean()
ax2.plot(mean_pptr.index, mean_hurdle.values,c="r")
plt.show()


testdf = pd.DataFrame({'ndatehour':Xtest[:,2].flatten()*1000,'pptr':Ytest.flatten(),'hurdle':hurdle_results['test_pred_hurdle_comb'].flatten()})
mpl.rcParams['figure.figsize'] = (16,4)
fig, ax = plt.subplots(1, 1, sharex=True)

mean_pptr = testdf.groupby('ndatehour')['pptr'].mean()
ax.bar(mean_pptr.index, mean_pptr.values, align='center')
ax2 = ax.twinx()
mean_hurdle = testdf.groupby('ndatehour')['hurdle'].mean()
ax2.plot(mean_pptr.index, mean_hurdle.values,c="r")
plt.show()


```
