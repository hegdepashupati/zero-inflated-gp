```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pickle
import glob
import pandas as pd
```

```python
rundir = "../models/run01_hz/1/"
```

```python
data = pickle.load(open(rundir+"data.pickle","rb"))
Xtrain = data['Xtrain']
Ytrain = data['Ytrain']
Xtest = data['Xtest']
Ytest = data['Ytest']
```

```python
clf_results = pickle.load(open(rundir+"results_scgp.pickle","rb"))
reg_results = pickle.load(open(rundir+"results_svgp.pickle","rb"))
```

```python
train_clf_prob = clf_results['pred_train']['pfmean']
test_clf_prob = clf_results['pred_test']['pfmean']
train_clf_indc = (train_clf_prob > 0.5) * 1.0
test_clf_indc = (test_clf_prob > 0.5) * 1.0
```

```python
train_zi_prob = train_clf_prob * reg_results['pred_train']['fmean']
test_zi_prob = test_clf_prob * reg_results['pred_test']['fmean']
train_zi_indc = train_clf_indc * reg_results['pred_train']['fmean']
test_zi_indc = test_clf_indc * reg_results['pred_test']['fmean']
```

```python
def rmse(predict,actual):
    predict = np.maximum(predict,0)
    return np.sqrt(np.mean((actual-predict)**2))

def mad(predict,actual):
    predict = np.maximum(predict,0)
    return np.mean(np.abs(actual-predict))

train_zi_prob_reg_rmse = rmse(train_zi_prob,Ytrain)
print("rmse on train set for zi svgp : "+str(train_zi_prob_reg_rmse))
train_zi_prob_reg_mae = mad(train_zi_prob,Ytrain)
print("mad on train set for zi svgp : "+str(train_zi_prob_reg_mae))

test_zi_prob_reg_rmse = rmse(test_zi_prob,Ytest)
print("rmse on test set for zi svgp  : "+str(test_zi_prob_reg_rmse))
test_zi_prob_reg_mae = mad(test_zi_prob,Ytest)
print("mad on test set for zi svgp  : "+str(test_zi_prob_reg_mae))

train_zi_indc_reg_rmse = rmse(train_zi_indc,Ytrain)
print("rmse on train set for zi svgp : "+str(train_zi_indc_reg_rmse))
train_zi_indc_reg_mae = mad(train_zi_indc,Ytrain)
print("mad on train set for zi svgp : "+str(train_zi_indc_reg_mae))

test_zi_indc_reg_rmse = rmse(test_zi_indc,Ytest)
print("rmse on test set for zi svgp  : "+str(test_zi_indc_reg_rmse))
test_zi_indc_reg_mae = mad(test_zi_indc,Ytest)
print("mad on test set for zi svgp  : "+str(test_zi_indc_reg_mae))
```

```python
# check plotting temporally - time on x axis, 
Xtrain = data['Xtrain']
Ytrain = data['Ytrain']
Xtest = data['Xtest']
Ytest = data['Ytest']
traindf = pd.DataFrame({'ndatehour':Xtrain[:,2].flatten()*1000,'pptr':Ytrain.flatten(),'probs':clf_results['pred_train']['pfmean'].flatten()})

mpl.rcParams['figure.figsize'] = (16,4)
fig, ax = plt.subplots(1, 1, sharex=True)

mean_pptr = traindf.groupby('ndatehour')['pptr'].mean()
ax.bar(mean_pptr.index, mean_pptr.values, align='center')
ax2 = ax.twinx()
mean_probs = traindf.groupby('ndatehour')['probs'].mean()
ax2.plot(mean_pptr.index, mean_probs.values,c="r")

plt.show()


testdf = pd.DataFrame({'ndatehour':Xtest[:,2].flatten()*1000,'pptr':Ytest.flatten(),'probs':clf_results['pred_test']['pfmean'].flatten()})

mpl.rcParams['figure.figsize'] = (16,4)
fig, ax = plt.subplots(1, 1, sharex=True)

mean_pptr = testdf.groupby('ndatehour')['pptr'].mean()
ax.bar(mean_pptr.index, mean_pptr.values, align='center')
ax2 = ax.twinx()
mean_probs = testdf.groupby('ndatehour')['probs'].mean()
ax2.plot(mean_pptr.index, mean_probs.values,c="r")

plt.show()

```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("train set:")
print(classification_report(Ytrain > 0 * 1,(clf_results['pred_train']['pfmean'] > 0.5)* 1))
# print(confusion_matrix(Ytrain > 0 * 1,(clf_results['pred_train']['pfmean'] > 0.5)* 1))

print("test set:")
print(classification_report(Ytest > 0 * 1,(clf_results['pred_test']['pfmean'] > 0.5)* 1))
# print(confusion_matrix(Ytest > 0 * 1,(clf_results['pred_test']['pfmean'] > 0.5)* 1))
```
