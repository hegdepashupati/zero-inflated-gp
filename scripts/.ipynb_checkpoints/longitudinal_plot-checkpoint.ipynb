```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pickle
import glob
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from copy import deepcopy
import matplotlib.gridspec as gridspec
import matplotlib as mpl
```

```python
data = pickle.load(open("../models/run01/1/data.pickle","rb"))
```

```python
timevale_list = [4.614,4.620,4.624,4.628] # ,4.632, 4.636] #range(4624,4636)
# timevale_list = [t/1000 for t in timevale_list]
```

```python
tf.reset_default_graph()
parentDir = "/l/hegdep1/onoffgp/uai/experiments/pptr/"
# parentDir = '/mnt/Documents/remote/cs256/onoffgp/uai/experiments/pptr'
sys.path.append(parentDir)
from onofftf.onoffpred import predict_onoff
```

```python
Xfull = np.vstack([data['Xtrain'],data['Xtest']])
Yfull = np.vstack([data['Ytrain'],data['Ytest']])

plotdict = {}

for timevalue in timevale_list:
    tind = np.where(Xfull[:,2] == timevalue)
#     X1actual = Xfull[tind,0][0].reshape(-1,1)
#     X2actual = Xfull[tind,0][0].reshape(-1,1)
    Yactual  = Yfull[tind]
    
    ngrid = 50
    # create basemap instance
    pmap = Basemap(resolution='l', # c, l, i, h, f or None
            projection='merc',
            lat_0=65.0, lon_0=25.0,
            llcrnrlon=19.08, llcrnrlat= 59.45, urcrnrlon=31.59, urcrnrlat=70.09)

    X1plot = []
    X2plot =[]
    for lat,lon in Xfull[tind,:2][0]:
        x, y = pmap(lon,lat)
        X1plot.append(x);X2plot.append(y)

    # prediction on test data
    Ypred =  predict_onoff(Xtrain = Xfull[tind,:3][0],Xtest=None,checkpointPath =parentDir+"/models/run_kronoff/model.ckpt")["gfmean"]

    # prediction on grid data
    lonsgrid,latsgrid = pmap.makegrid(ngrid,ngrid)
    xgrid, ygrid = pmap(lonsgrid, latsgrid)
    griddata = np.array([latsgrid.flatten(),
                         lonsgrid.flatten(),
                         np.repeat(timevalue,ngrid*ngrid)])
    griddata = griddata.T
    
    Ppred = predict_onoff(Xtrain = griddata,Xtest=None,checkpointPath = parentDir+"/models/run_kronoff/model.ckpt")
    
    plotdict[str(timevalue)] = {'X1plot':X1plot,
                                'X2plot':X2plot,
                                'Yactual':Yactual,
                                'Ypred':Ypred,
                                'xgrid':xgrid,
                                'ygrid':ygrid,
                                'gfmean':Ppred['gfmean'],
                                'fmean':Ppred['fmean'],
                                'pgmean':Ppred['pgmean']}
```

```python
from matplotlib import ticker
import matplotlib.colors as colors

tick_locator = ticker.MaxNLocator(nbins=4)

mpl.rcParams['figure.figsize'] = (9,12)
mpl.rcParams.update({'font.size': 18})
norm = mpl.colors.Normalize(vmin=-0.3,vmax=2.5)
fig = plt.figure()
gs = gridspec.GridSpec(3,len(timevale_list))

for col in range(len(timevale_list)):
    X1plot = plotdict[str(timevale_list[col])]['X1plot']
    X2plot = plotdict[str(timevale_list[col])]['X2plot']
    Yactual = plotdict[str(timevale_list[col])]['Yactual']
    Ypred = plotdict[str(timevale_list[col])]['Ypred']
    xgrid = plotdict[str(timevale_list[col])]['xgrid']
    ygrid = plotdict[str(timevale_list[col])]['ygrid']
    gfmean = plotdict[str(timevale_list[col])]['gfmean']
    fmean = plotdict[str(timevale_list[col])]['fmean']
    pgmean = plotdict[str(timevale_list[col])]['pgmean']
    
    ax1 = fig.add_subplot(gs[0,col])
    p1 = deepcopy(pmap)
    p1.drawcountries(linewidth=0.8)
    p1.drawcoastlines(linewidth=0.4,linestyle="--")
    pp = p1.scatter(X1plot,X2plot,c=Yactual.flatten(),norm=norm,alpha=0.85,s=30, marker='o',cmap='RdBu_r')
    if col == 3:
        cbar = p1.colorbar(extend='max')
        cbar.locator = tick_locator
        cbar.update_ticks()


    # plot 03
    ax3 = fig.add_subplot(gs[1,col])
    p3 = deepcopy(pmap)
    p3.drawcountries(linewidth=0.4)
    p3.drawcoastlines(linewidth=0.2,linestyle="--")
    p3.contourf(xgrid,ygrid,gfmean.reshape(ngrid,ngrid),cmap='RdYlGn',alpha=0.7)
    if col == 3:
        cbar = p3.colorbar(extend='max')
        cbar.locator = tick_locator
        cbar.update_ticks()

    # plot 04
    ax4 = fig.add_subplot(gs[2,col])
    p4 = deepcopy(pmap)
    p4.drawcountries(linewidth=0.4)
    p4.drawcoastlines(linewidth=0.2,linestyle="--")
    p4.contourf(xgrid,ygrid,pgmean.reshape(ngrid,ngrid),cmap='Blues',alpha=0.7)
    if col == 3:
        cbar = p4.colorbar(extend='max')
        cbar.locator = tick_locator
        cbar.update_ticks()


# gs.tight_layout(fig)
plt.figtext(0.5,0.9, "(a) Data samples", ha="center", va="center", fontsize=18)
plt.figtext(0.5,0.636, "(b) Sparse latent function  "+ r"$\mathbf{f}|\mathbf{g}$", ha="center", va="center", fontsize=18)
plt.figtext(0.5,0.362, "(c) Support function  " + r"$\Phi(\mathbf{g})$", ha="center", va="center", fontsize=18)
plt.subplots_adjust(hspace = 0.2,wspace=0.01)
plt.savefig("02_02_pptr.png")
plt.show()

```
